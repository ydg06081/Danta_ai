```python
import pandas as pd

def read_excel_and_save_csv(xlsx_path, csv_path):
    """
    주어진 엑셀 파일(.xlsx)을 읽어서 pandas DataFrame으로 반환하고, 같은 데이터를 csv 파일로 저장합니다.
    """
    try:
        df = pd.read_excel(xlsx_path)
        print(f"파일에서 읽은 데이터:\n{df.head()}")
        df.to_csv(csv_path, index=False, encoding='utf-8-sig')
        print(f"CSV 파일로 저장되었습니다: {csv_path}")
        return df
    except Exception as e:
        print(f"엑셀 파일 읽기 또는 CSV 저장 오류: {e}")
        return None


df = read_excel_and_save_csv("NewsResult_20241001-20251001.xlsx", "news.csv")

```
